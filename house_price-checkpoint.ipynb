# House Price data analysis project

importing libraries 

```python
import pandas as pd          # For handling data in DataFrames
import numpy as np           # For numerical computations
import matplotlib.pyplot as plt  # For creating plots
import seaborn as sns

```

Create the dataframe 

```python
dataframe = pd.read_csv("House_Price.csv")
print(dataframe)
```

```python
dataframe
```

summary of dataframe 

```python
dataframe.info()
```

there is no null value

```python
# Check for missing values
print(dataframe.isnull().sum())
```

```python
dataframe['n_hos_beds'] = dataframe['n_hos_beds'].fillna(dataframe['n_hos_beds'].mean())

```

```python
dataframe['waterbody'] = dataframe['waterbody'].fillna('None')
```

```python
print(dataframe.isnull().sum())
```

```python
# Rename columns for consistency
dataframe.columns = dataframe.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
print(dataframe.dtypes)

# Convert categorical columns
categorical_columns = ['resid_area', 'waterbody', 'airport', 'bus_ter']
dataframe[categorical_columns] = dataframe[categorical_columns].astype('category')
```

```python
# Check for duplicates
print(dataframe.duplicated().sum())

# Drop duplicates if any
dataframe.drop_duplicates(inplace=True)
```

```python
dataframe.describe(include="all")
```

```python
# Convert categorical columns
dataframe['room_num'] = dataframe['room_num'].astype('category')

```

```python
sns.histplot(dataframe['price'], kde=True, bins=30)
plt.title("Distribution of House Prices")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.histplot(dataframe['crime_rate'], kde= True , bins=10, edgecolor='black')
plt.title("Crime rate distribution")
plt.xlabel("Crime Rate")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.boxplot(dataframe['crime_rate'])
plt.title("Crime Rate Boxplot")
plt.xlabel("Crime Rate")
plt.show()
```

```python
sns.countplot(dataframe['room_num'])
plt.title("Room Count Frequency")
plt.xlabel("Number of Rooms")
plt.ylabel("Count")
plt.show()
```

```python
sns.scatterplot(data=dataframe, x='room_num', y='price')
plt.title("price VS no.of rooms")
plt.xlabel("Room number")
plt.ylabel("Price")
plt.show()

```

```python
sns.scatterplot(data= dataframe,x= 'price', y= 'crime_rate')
plt.title("Price vs Crime rate")
plt.xlabel("Price")
plt.ylabel("crime rate")
plt.show()
```

```python
sns.scatterplot(data=dataframe, x='dist1', y='price')
plt.title("Price vs. Distance to district 1")
plt.xlabel("Distance to district1")
plt.ylabel("Price")
plt.show()
```

```python
sns.scatterplot(data=dataframe, x='dist2', y='price')
plt.title("Price vs. Distance to district 2")
plt.xlabel("Distance to District 2")
plt.ylabel("Price")
plt.show()
```

```python
sns.scatterplot(data=dataframe, x='dist3', y='price')
plt.title("Price vs. Distance to district 3")
plt.xlabel("Distance to District 3")
plt.ylabel("Price")
plt.show()
```

```python
numeric_data = dataframe.select_dtypes(include=['number'])
correlation_matrix = numeric_data.corr()
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
sns.heatmap(dataframe.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Values Heatmap')
plt.show()
```

conclusion
•  House Prices:
•	The distribution of house prices is positively skewed, with most homes priced within a mid-range.
•	High-priced outliers suggest the presence of luxury properties, which may require special treatment in predictive modeling to avoid distortion.
•  Crime Rate:
•	The crime rate is heavily skewed, with several areas exhibiting extremely high crime rates.
•	These outliers could impact the model’s ability to generalize and may require capping or transformation.
•  Room Numbers and Price:
•	There appears to be a positive relationship between the number of rooms (room_num) and house prices, as seen in the scatter plots or pair plot.
•	This relationship suggests that larger homes generally command higher prices.
•  Negative Factors:
•	Features such as crime_rate and poor_prop are likely negatively correlated with house prices, indicating that higher crime rates or higher proportions of lower-income households adversely affect property values.
•  Outliers:
•	Significant outliers are present in key variables such as price, crime_rate, and possibly others. These outliers may skew model performance and need to be addressed during preprocessing.
