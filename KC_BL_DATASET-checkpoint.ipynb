# kc bl dataset

importing libraries 

```python
import pandas as pd          # For handling data in DataFrames
import numpy as np           # For numerical computations
import matplotlib.pyplot as plt  # For creating plots
import seaborn as sns
```

Create the dataframe

```python
dataframe = pd.read_csv("KC_BL_DATASET.csv")
print(dataframe)
```

```python
dataframe
```

summary of dataframe

```python
dataframe.info()
```

```python
print(dataframe.isnull().sum())
```

```python
dataframe['Genre_2'] = dataframe['Genre_2'].fillna('None')
```

```python
dataframe['Genre_3'] = dataframe['Genre_3'].fillna('None')
```

```python
print(dataframe.isnull().sum())
```

```python
# Rename columns for consistency
dataframe.columns = dataframe.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
print(dataframe.dtypes)

# Convert categorical columns
categorical_columns = ['main_genre', 'genre_2', 'genre_3','studio','title']
dataframe[categorical_columns] = dataframe[categorical_columns].astype('category')
```

```python
# Check for duplicates
print(dataframe.duplicated().sum())

# Drop duplicates if any
dataframe.drop_duplicates(inplace=True)
```

```python
dataframe.describe(include="all")
```

```python
sns.histplot(dataframe['price'], bins=10, kde=True)
plt.title("Distribution of Price")
plt.xlabel("price")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.scatterplot(x='bedrooms', y='price', data=dataframe)
plt.title("Price vs. Number of Bedrooms")
plt.xlabel("Number ofBedrooms")
plt.ylabel("Price")
plt.show()
```

```python
sns.scatterplot(data=dataframe, x='sqft_living', y='price')
plt.title('Price vs. square footage(living room)')
plt.xlabel('Suarefootage(living room)')
plt.ylabel('Price')
plt.show()
```

```python
numeric_data = dataframe.select_dtypes(include=['number'])
correlation_matrix = numeric_data.corr()
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()

```

```python
sns.lineplot(x='price', y='yr_built',data=dataframe)
plt.title("Price VS Year Built")
plt.xlabel("Price")
plt.ylabel("Year Built")
plt.show()
```

```python
sns.scatterplot(data=dataframe, x='yr_built', y='price')
plt.title('Price vs. Year built')
plt.xlabel('Year built')
plt.ylabel('Price')
plt.show()
```

```python
sns.scatterplot(data=dataframe, x='imdb_rating', y='price')
plt.title('Price vs. Rating')
plt.xlabel('Rating')
plt.ylabel('Price')
plt.show()
```

```python
sns.boxplot(data=dataframe, x='genre_2', y='price')
plt.title('Price Distribution by Genre')
plt.xlabel('Genre')
plt.ylabel('Price')
plt.show()
```

```python
sns.heatmap(dataframe.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Values Heatmap')
plt.show()
```

```python
sns.violinplot(x='studio', y='price', data=dataframe)
plt.title("Violin Plot studio VS price")
plt.xlabel("Studio")
plt.ylabel("Price")
plt.show()
```

```python
sns.pairplot(dataframe[['bedrooms', 'bathrooms', 'price']])
plt.title('Pairplot: Bedrooms, Bathrooms, and Price')
plt.show()
```

Conclusion
1.	Price Distribution:
o	Histogram/Kernel Density Estimate (KDE): Price is right-skewed; most properties are below a specific price threshold.
2.	Location-Based Trends:
o	Bar Chart: Average property price by Location. Some locations show much higher mean prices.
o	Heatmaps/Choropleths: Geographic distribution of prices (if location coordinates are included).
3.	Year Built vs. Price:
o	Older homes tend to have a wider price range but may not always be the most expensive.
o	Trendline shows no consistent increase in price for recently built properties.
4.	Living Area vs. Price:
o	Larger properties generally have higher prices, but diminishing returns observed after a certain size.
5.	Bedrooms/Bathrooms Distribution:
o	Properties with 2-4 bedrooms dominate the dataset.
o	Number of bathrooms aligns closely with the number of bedrooms.
•	Skewed Data: Several features, especially Price, require transformations or outlier handling if modeling is intended.
•	Important Predictors: Features like Living Area, Location, and Bedrooms are strong predictors of price.
